```python
# Load the source database
# Asign headers and build the dataframe

import pandas as pd
headers=["Title", "Author", "Pages", "Year", "Genres", "Rating", "Avr", "Link"]
df = pd.read_csv('goodreadsdb.csv', names=headers)
```

```python
# Search for no title books

df.loc[(df['Title'] == "No encontrado") | (df['Title'] == ",,")]
```

```python
# Search for no author books

df.loc[(df['Author'] == "No encontrado") | (df['Author'] == ",,")]
```

```python
# Search for no pages books

df.loc[(df['Pages'] == "No encontrado") | (df['Pages'] == ",,")]
```

```python
# Search for no published date books

df.loc[(df['Year'] == "No encontrado") | (df['Year'] == ",,")]
```

```python
# Get the no pages books to correct

nopages = df.loc[(df['Pages'] == "No encontrado") | (df['Pages'] == ",,")]
nopages.to_csv('nopages.csv', index=False, header=False)
```

```python
# Get the no year books to correct

noyear = df.loc[(df['Year'] == "No encontrado") | (df['Year'] == ",,")]
noyear.to_csv('noyear.csv', index=False, header=False)
```
