```python
import pandas as pd

# import the database and asign column names

headers=["Title", "Author", "Pages", "Year", "Genres", "Rating", "Avr", "Link"]
df = pd.read_csv('database.csv', names=headers)
```

```python
# year as integer

df['Year'] = pd.to_numeric(df["Year"], errors='coerce', downcast='integer')
```

```python
import matplotlib.pyplot as plt
```

```python
# filter books 

dis_filt = df.loc[(df['Year'] > 1700) & (df['Genres'].apply(lambda x: 'dystopia' in x))]
dis_data = dis_filt['Year'].value_counts().rename_axis('Year').reset_index(name='Books').sort_values('Year')

class_filt = df.loc[(df['Year'] > 1700) & (df['Genres'].apply(lambda x: 'classics' in x))]
class_data = class_filt['Year'].value_counts().rename_axis('Year').reset_index(name='Books').sort_values('Year')

plt.plot(dis_data['Year'], dis_data['Books'], label='dystopia')
plt.plot(class_data['Year'], class_data['Books'], label='classics')

plt.grid()
plt.legend(loc='upper left')
plt.title('Libros indexados por tags', fontweight='bold')
plt.ylabel('Cantidad')
plt.xlabel('Año de publicación')
plt.savefig('images/dysversusclass.png')
plt.show()
```

```python
# filter books 

dis_filt2 = df.loc[(df['Year'] > 1990) & (df['Genres'].apply(lambda x: 'dystopia' in x))]
dis_data2 = dis_filt2['Year'].value_counts().rename_axis('Year').reset_index(name='Books').sort_values('Year')

chris_filt = df.loc[(df['Year'] > 1990) & (df['Genres'].apply(lambda x: 'christian' in x))]
chris_data = chris_filt['Year'].value_counts().rename_axis('Year').reset_index(name='Books').sort_values('Year')

plt.plot(dis_data2['Year'], dis_data2['Books'], label='dystopia')
plt.plot(chris_data['Year'], chris_data['Books'], label='christian')

plt.grid()
plt.legend(loc='upper left')
plt.title('Libros indexados por tags', fontweight='bold')
plt.ylabel('Cantidad')
plt.xlabel('Año de publicación')
plt.savefig('images/dysversuschrist.png')
plt.show()
```
