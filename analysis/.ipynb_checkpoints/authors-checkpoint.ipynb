```python
import pandas as pd

# import the database and asign column names

headers=["Title", "Author", "Pages", "Year", "Genres", "Rating", "Avr", "Link"]
df = pd.read_csv('database.csv', names=headers)
```

```python
top20authors = (df['Author'].value_counts().head(20)).rename_axis('Author').reset_index(name='Books')
top20_df = pd.DataFrame(top20authors)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(top20_df['Author'],top20_df['Books'])
plt.title('Autores más indexados', fontweight='bold')
plt.xlabel('Cantidad de libros')
plt.savefig('images/top20authors.png')
plt.show()
```

```python
# filter the books of the top 20

top20books = df.loc[(df['Author'].isin(top20_df['Author'].tolist()))]
```

```python
top20books.head()
```

```python
tags = ((top20books.Genres.apply(lambda x: x[1:-1].strip("[]").replace("'", "").replace(" ", "").split(","))).explode().unique()).tolist()
```

```python
records = []

for tag in tags:
    if tag == "": pass
    else:
        occurrences = len([x for x in top20books['Genres'] if tag in x])
        data = [tag, occurrences]
        records.append(data)
```

```python
top20tags = pd.DataFrame(records, columns = ['Tag', 'Occurrences'])
```

```python
top20tags.sort_values(by=['Occurrences'], ascending=False).head()
to_deploy = top20tags.sort_values(by=['Occurrences'], ascending=False).head(15)
```

```python
my_data = to_deploy['Occurrences']
my_labels = to_deploy['Tag']
plt.pie(my_data,labels=my_labels,autopct='%1.1f%%')
plt.title('Etiquetas más utilizadas por los veinte autores más populares', fontweight='bold')
plt.axis('equal')
plt.savefig('images/top20authorstags.png')
plt.show()
```
