```python
import pandas as pd

# import the database and asign column names

headers=["Title", "Author", "Pages", "Year", "Genres", "Rating", "Avr", "Link"]
df = pd.read_csv('database.csv', names=headers)
```

```python
# year as integer

df['Year'] = pd.to_numeric(df["Year"], errors='coerce', downcast='integer')
```

```python
# filter books published after 1500

after1500 = df.loc[(df['Year'] > 1500)]
```

```python
# how many books were published by year

countbooks = after1500['Year'].value_counts().rename_axis('Year').reset_index(name='Books').sort_values('Year')
```

```python
countbooks.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(countbooks['Year'], countbooks['Books'])
plt.title('Goodreads Best Books Ever', fontweight='bold')
plt.ylabel('Libros indexados')
plt.xlabel('A침o de publicaci칩n')
plt.grid()
plt.savefig('images/booksafter1500.png')
plt.show()
```

```python
after1900 = df.loc[(df['Year'] > 1900)]
countbooks2 = after1900['Year'].value_counts().rename_axis('Year').reset_index(name='Books').sort_values('Year')

plt.plot(countbooks2['Year'], countbooks2['Books'])
plt.title('Goodreads Best Books Ever', fontweight='bold')
plt.ylabel('Libros indexados')
plt.xlabel('A침o de publicaci칩n')
plt.grid()
plt.savefig('images/booksafter1900.png')
plt.show()
```
